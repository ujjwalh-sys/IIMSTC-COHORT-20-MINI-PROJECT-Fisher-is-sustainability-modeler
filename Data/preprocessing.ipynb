```python
import pandas as pd
import os
```

```python
df = pd.read_csv("final_fisheries_dataset.csv")

print("Original Shape:", df.shape)
df.head()
```

```python
df = df.drop_duplicates()

print("After Removing Duplicates:", df.shape)
```

```python
df = df.drop_duplicates()

# Fix biomass column
df["biomass_relative_to"] = (
    df["biomass_relative_to"]
    .astype(str)
    .str.replace(".", "", regex=False)
)

df["biomass_relative_to"] = pd.to_numeric(df["biomass_relative_to"], errors="coerce")

# Fill only numeric columns
df = df.fillna(df.select_dtypes(include="number").mean())
```

```python
df["year"] = df["year"].astype(int)
df["SST"] = df["SST"].astype(float)
df["Trawling_Hours"] = df["Trawling_Hours"].astype(float)
df["biomass_relative_to"] = df["biomass_relative_to"].astype(float)

df.dtypes
```

```python
output_path = "preprocessed_fisheries_dataset.csv"
df.to_csv(output_path, index=False)

print("✅ Preprocessed file saved at:")
print(os.path.abspath(output_path))
```
