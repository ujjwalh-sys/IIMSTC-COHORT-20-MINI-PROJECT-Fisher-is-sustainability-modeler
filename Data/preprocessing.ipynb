```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
df = pd.read_csv("final_fisheries_dataset.csv")
df.head()
```

```python

df["biomass_relative_to"] = (
    df["biomass_relative_to"]
        .astype(str)
        .str.replace(".", "", regex=False)
        .astype(float)
)


df["SST"] = df["SST"].fillna(df["SST"].mean())
df["Trawling_Hours"] = df["Trawling_Hours"].fillna(df["Trawling_Hours"].mean())
df["biomass_relative_to"] = df["biomass_relative_to"].fillna(
    df["biomass_relative_to"].mean()
)

df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
le = LabelEncoder()
df["country_encoded"] = le.fit_transform(df["country"])
df.head()
```

```python


scaler = StandardScaler()

features = ["year", "SST", "Trawling_Hours"]
df[features] = scaler.fit_transform(df[features])
```

```python
df.to_csv("preprocessed_fisheries_dataset.csv", index=False)

print("Preprocessed CSV file created successfully!")
df.head()
```
